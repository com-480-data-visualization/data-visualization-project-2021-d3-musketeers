```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy, copy
```

```python
titles = pd.read_csv('../../data/titles.csv')
titles['release date on Netflix'] = pd.to_datetime(titles['release date on Netflix'])

movies = titles[titles['duration'].notna()]
shows = titles[titles['number of seasons'].notna()]
```

```python
movies_age_rating_release = movies[['age rating', 'release date on Netflix']]
shows_age_rating_release = shows[['age rating', 'release date on Netflix']]
# titles_age_rating_release
```

```python
shows
```

```python
# age_rating_stats = pd.DataFrame(columns=['age_rating', 'year', 'percentage'])

stats_serieses = []
for category_name, category_frame in [('movies', movies_age_rating_release), ('shows', shows_age_rating_release)]:
    for stat_year in range(2015, 2022):
        titles_up_to_year = category_frame[category_frame['release date on Netflix'].dt.year <= stat_year]
        age_rating_percentage_up_to_year = titles_up_to_year.groupby('age rating').count() / titles_up_to_year.count()
        age_rating_percentage_up_to_year['year'] = stat_year
        age_rating_percentage_up_to_year.columns = ['', 'percentage', 'year']
        age_rating_percentage_up_to_year = age_rating_percentage_up_to_year[['year', 'percentage']]
        age_rating_percentage_up_to_year.reset_index()
        age_rating_percentage_up_to_year = age_rating_percentage_up_to_year[age_rating_percentage_up_to_year.percentage > 0.01]
        age_rating_percentage_up_to_year['category'] = category_name
        stats_serieses.append(age_rating_percentage_up_to_year)
    
age_rating_per_year_distribution = pd.concat(stats_serieses, axis=0)
age_rating_per_year_distribution
age_rating_per_year_distribution.to_csv('age_rating_per_year_distribution.csv')
```

# Density of releases

```python
titles_release_month = titles[['title', 'release date on Netflix']]

titles_release_month['release date on Netflix'] = titles_release_month['release date on Netflix']\
    .map(lambda date: date.replace(day=1))

titles_release_month
```

```python
releases_per_month = titles_release_month.groupby('release date on Netflix').count()
releases_per_month = releases_per_month.reset_index()
releases_per_month.columns = [['release month', 'count']]

release_dates = releases_per_month['release month']
min_date = release_dates.min()[0]
max_date = release_dates.max()[0]

current_date = min_date

# missing_dates = pd.DataFrame(columns=['release month', 'count'])

missing_dates = []
while current_date < max_date:
    
    is_in = (release_dates == current_date).any().bool()
#     print(current_date, is_in)
    if not is_in:
        missing_dates.append(copy(current_date))
    current_date += pd.DateOffset(months=1)
    
missing_dates_df = pd.DataFrame([[date, 0] for date in missing_dates], columns=[('release month',), ('count',)])
releases_per_month = pd.concat([releases_per_month, missing_dates_df])

releases_per_month.to_csv('release_density_per_month.csv', date_format='%m-%d-%Y')
```
